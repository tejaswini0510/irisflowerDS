```python
#importing all the essential libraries


import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
```

loading dataset


```python
columns = ['sepal length','sepal width','petal length','petal width','class_lables']
# load the data
df = pd.read_csv('irisdataset.txt', names=columns)
df.head()
```

visualization of dataset

```python
df.describe()
```

```python
#visualize the whole dataset
sns.pairplot(df,hue='class_lables')
```

seperating input columns and output columns


```python
#seperate features and target
data=df.values

X=data[:,0:4]
Y=data[:,4]
print(X)
print(Y)
```

SPLITTING THE DATA INTO TRAINING AND TESTING

```python
#split the data to train and test dataset.
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
print(y_train)
```

MODEL1:support vector machine algorithm

```python
#support vector machine algorithm
from sklearn.svm import SVC

model_svc=SVC()
model_svc.fit(X_train,y_train)
```

```python
prediction1=model_svc.predict(X_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction1)*100)
for i in range(len(prediction1)):
  print(y_test[i],prediction1[i])
```

MODEL2:LOGISTIC REGRESSION

```python
#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train,y_train)
```

```python
prediction2=model_LR.predict(X_test)
#calculate the accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction2)*100)
for i in range(len(prediction1)):
  print(y_test[i],prediction1[i])
```
